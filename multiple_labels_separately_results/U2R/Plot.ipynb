```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D as axes3d
import numpy as np
from random import randrange
```

```python
label = "U2R"
```

```python
df_DT = pd.read_csv(label+'_DT.csv')
```

```python
print(df_DT)
```

```python
df_DT_feature_selection = df_DT.groupby(['feature_selection'])

fig, ax = plt.subplots()
df_DT_feature_selection.get_group(' PCA').plot(kind='scatter', x='depth', y='f1_scores(mean)',
                                               s=60, c='percentage_selection', cmap='RdYlGn'
                                               ,ax=ax,title="PCA")
fig2,ax2 = plt.subplots()
df_DT_feature_selection.get_group(' ANOVA').plot(kind='scatter', x='depth', y='f1_scores(mean)',
                                               s=60, c='percentage_selection', cmap='RdYlGn'
                                               ,ax=ax2,title="ANOVA")
fig.savefig(label+"_DT"+"_PCA"+"_F1"+".png", dpi=500)
fig2.savefig(label+"_DT"+"_ANOVA"+"_F1"+".png", dpi=500)

fig3, ax3 = plt.subplots()
df_DT_feature_selection.get_group(' PCA').plot(kind='scatter', x='depth', y='roc_auc(mean)',
                                               s=60, c='percentage_selection', cmap='RdYlGn'
                                               ,ax=ax3,title="PCA")
fig4,ax4 = plt.subplots()
df_DT_feature_selection.get_group(' ANOVA').plot(kind='scatter', x='depth', y='roc_auc(mean)',
                                               s=60, c='percentage_selection', cmap='RdYlGn'
                                               ,ax=ax4,title="ANOVA")
fig3.savefig(label+"_DT"+"_PCA"+"_ROC"+".png", dpi=500)
fig4.savefig(label+"_DT"+"_ANOVA"+"_ROC"+".png", dpi=500)
```

```python
df_LG = pd.read_csv(label+'_LG.csv')
df_LG_feature_selection = df_LG.groupby(['feature_selection'])

fig, ax = plt.subplots()
df_LG_feature_selection.get_group(' PCA').plot(kind='scatter', x='solver', y='f1_scores(mean)',
                                               s=60, c='percentage_selection', cmap='RdYlGn'
                                               ,ax=ax,title="PCA")
fig2,ax2 = plt.subplots()
df_LG_feature_selection.get_group(' ANOVA').plot(kind='scatter', x='solver', y='f1_scores(mean)',
                                               s=60, c='percentage_selection', cmap='RdYlGn'
                                               ,ax=ax2,title="ANOVA")
fig.savefig(label+"_LG"+"_PCA"+"_F1"+".png", dpi=500)
fig2.savefig(label+"_LG"+"_ANOVA"+"_F1"+".png", dpi=500)

fig3, ax3 = plt.subplots()
df_LG_feature_selection.get_group(' PCA').plot(kind='scatter', x='solver', y='roc_auc(mean)',
                                               s=60, c='percentage_selection', cmap='RdYlGn'
                                               ,ax=ax3,title="PCA")
fig4,ax4 = plt.subplots()
df_LG_feature_selection.get_group(' ANOVA').plot(kind='scatter', x='solver', y='roc_auc(mean)',
                                               s=60, c='percentage_selection', cmap='RdYlGn'
                                               ,ax=ax4,title="ANOVA")
fig3.savefig(label+"_LG"+"_PCA"+"_ROC"+".png", dpi=500)
fig4.savefig(label+"_LG"+"_ANOVA"+"_ROC"+".png", dpi=500)
```

```python
df_SVM = pd.read_csv(label+'_SVM.csv')
df_SVM_feature_selection = df_SVM.groupby(['kernel_selection','feature_selection'])

fig, ax = plt.subplots()
tuple_selection = (' poly', ' PCA')
df_SVM_feature_selection.get_group(tuple_selection)[(df_SVM_feature_selection.get_group(tuple_selection)['C_selection']<10)].plot(kind='scatter', x='percentage_selection', y='f1_scores(mean)',use_index=True,
                                               s=60, c='C_selection', cmap='RdYlGn'
                                                ,ax=ax,title="ploy PCA")
tuple_selection = (' poly', ' ANOVA')
fig2,ax2 = plt.subplots()
df_SVM_feature_selection.get_group(tuple_selection)[(df_SVM_feature_selection.get_group(tuple_selection)['C_selection']<10)].plot(kind='scatter', x='percentage_selection', y='f1_scores(mean)',use_index=True,
                                               s=60, c='C_selection', cmap='RdYlGn'
                                               ,ax=ax2,title="ploy ANOVA")
fig.savefig(label+"_SVM"+"_PLOY_PCA"+"_F1"+".png", dpi=500)
fig2.savefig(label+"_SVM"+"_PLOY_ANOVA"+"_F1"+".png", dpi=500)
```
